```python
x=iris.data   #feature names
y=iris.target  #target names
```

```python
type(iris.target)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
xtrain.shape
```

```python
#classification -knn - k_Nearest_Neighbours
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
```

```python
knn.fit(xtrain,ytrain)
```

```python
knn.score(xtest,ytest)
```

```python
knn.predict([[3,2,4,6]])
```

```python
iris.target_names[knn.predict([[3,2,4,6],[3,4,5,6]])]
```

```python
# make a loop to calculate accuracy in range 0-15 and plot the values on x and y axis representing k-value vs accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
y=np.arange(15)
x=np.arange(15)
for k in range(1,16):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrain,ytrain)
    x[k-1]=k
    y[k-1]=knn.score(xtest,ytest)*100
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel('Values of k')
plt.ylabel('Accuracy')
plt.show()
```
